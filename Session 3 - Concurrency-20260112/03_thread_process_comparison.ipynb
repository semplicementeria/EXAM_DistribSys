# What happens when we increase the number of parallel processes?

For this section of the laboratory you need a Linux-based machine

Upload/copy the `multi-proc_thread.py` script on the Linux machine

Open three terminal windows.

In terminal 1, obtain the current bash interpreter process ID with `echo $$`

In terminal 2, monitor the process tree starting from the bash interpreter on terminal 1 with `watch -d -n 1 pstree <term1 PID>`, specifying the process ID obtained before

In terminal 3, run the `top` command

In terminal 1, launch the `multi-proc_thread.py` script with increasing parallelism [ 1, 2, 4] (`-n` option)

Limit the maximum amount of CPU that your script can use running the following commands in terminal 1:

`echo $$ | sudo tee /sys/fs/cgroup/test/cgroup.procs`

`echo '100000 100000' | sudo tee /sys/fs/cgroup/test/cpu.max`

Then run the script again with option `-n 4`

Limit the execution of the script to a single CPU running the following command in terminal 1:

`echo '0' | sudo tee /sys/fs/cgroup/test/cpuset.cpus`

Then run the script again with option `-n 4`
