### Objective:
The objective of this exercise is to create a command-line tool that retrieves JSON data from a remote source, filters it based on user ID, stores the filtered data locally in YAML format, and allows the user to add new TODO items to the local copy.

### Requirements:
1. **Command-line Interface (CLI) Arguments Parsing**:
   - Parse the following arguments from the command line interface:
     - Remote endpoint URL (string, optional, default 'https://dummyjson.com/todos')
     - User ID (integer, optional, default 1)

2. **Retrieve Remote Data**:
   - Fetch JSON data from the specified remote endpoint URL.

3. **Filter Data by User ID**:
   - Filter the retrieved JSON data to keep only TODOs related to the specified user ID.

4. **Print Filtered Data**:
   - Print the filtered data to the console for the user to review.

5. **Save Filtered Data Locally**:
   - Save the filtered data locally as a YAML file.

6. **User Interaction**:
   - Prompt the user to insert a new TODO item [check input()](https://docs.python.org/3/library/functions.html#input).
      - Ask for a description of the new TODO.
   - Generate a unique TODO ID for it.
   - Save the new TODO item to the local file.

### Additional Notes:
- Ensure proper error handling for network requests and file operations.
- Validate user inputs, such as ensuring the user ID is an integer.
- Implement user-friendly messages and prompts to guide the user through the interaction process.
- (optional) Consider modularizing the code into functions for better readability and maintainability.
- (optional) Document the code with comments or docstrings to explain its functionality and usage.
