# Examples of socket programming in python

This practice is intended to be run using multiple terminal windows.

## 1. Datagram socket

Upload/copy the `udp-client.py` and `udp-server.py` scripts.

Launch the server script first. On Linux, you can check that the server is waiting to receive incoming data by typing `netstat -naup --ip` on another terminal window (if you are using Windows, try with `netstat -n -a -o`. Instead with MacOS try with `netstat -na -p udp`).

Then launch the client script from yet another terminal window and send some text.

Repeat the client request multiple times, paying attention to how the client socket address reported by the server changes.

## 2. Stream socket, iterative server

Upload/copy the `tcp-client.py` and `tcp-server-iter.py` scripts.

Launch the server script first. On Linux, you can check that the server is in the TCP `LISTEN` state waiting for incoming connection requests by typing `netstat -nat --ip` on another terminal window (if you are using Windows, try with `netstat -n -a -o`. Instead with MacOS try with `netstat -na -p tcp`).

Then launch the client script from yet another terminal window and send some text. You can check that the new connection is in the TCP `ESTABLISHED` using `netstat`.

Send multiple text strings from the client, paying attention to how the client socket address reported by the server does not change.

You can close the connection by sending an empty text string.

Repeat the client request multiple times, launching the script from separate terminals. Verify how each client instance is blocked until the previous ones have not been terminated.

<img src="stream-iter.png" alt="Stream socket primitives, iterative server" width="600"/>





## 3. Stream socket, concurrent server

Upload/copy the `tcp-server-conc.py` script, then launch it, choosing between multithreading or multiprocessing.

The server prints out its process ID (PID). On Linux, to check for any child process/thread to be spawned, launch the following command on another terminal: `watch -d -n 1 pstree -p ` \<PID\>.

Launch multiple client instances from separate terminals. Verify how each client instance is now not blocked by the previous ones. Check also how the output of the `pstree` command changes. You can also check that the new connections are in the TCP `ESTABLISHED` state using `netstat` on yet another terminal.
